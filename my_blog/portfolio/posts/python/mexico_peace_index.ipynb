## What does it measue?

- Analyzes recent violence and peace trends.
- Estimates the economic cost of violence in Mexico.

## How it’s measured

The MPI isn’t just about crime rates. It considers 12 sub-indicators grouped into five major categories:

- Ongoing Conflict
- Safety and Security
- Domesticized Conflict
- Militarization
- Incarceration
